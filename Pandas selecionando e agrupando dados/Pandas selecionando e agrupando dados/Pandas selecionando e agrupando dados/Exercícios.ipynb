```python
excel = '1-SEEG10_GERAL-BR_UF_2022.10.27-FINAL-SITE.xlsx'
emissoes_gases = pd.read_excel(excel, sheet_name = 'GEE Estados')
```

```python
# 2
emissoes_gases_sul = emissoes_gases[emissoes_gases["Estado"].isin(['RS', 'PR', 'SC'])]
emissoes_gases_sul
```

```python
#3
emissoes_gases_amazonas = emissoes_gases[(emissoes_gases['Nível 1 - Setor'] == 'Mudança de Uso da Terra e Floresta') & (emissoes_gases['Estado'] == 'AM')]
emissoes_gases_amazonas
```

```python
#4
print(emissoes_gases.loc[(emissoes_gases['Nível 1 - Setor'] == 'Agropecuária') & (emissoes_gases['Estado'] == 'PA'), 2021].max())
```

```python
# Exercícios aula 2
```

```python
emissoes_setor = emissoes_gases.groupby('Nível 1 - Setor').sum(numeric_only = True)
```

```python
emissoes_setor
```

```python
emissao_agropecuaria = emissoes_setor.iloc[0].sum()
emissao_agropecuaria
```

```python
colunas_emissao = list(emissoes_gases.loc[:, 1970:2021].columns)
colunas_info = list(emissoes_gases.loc[:, 'Nível 1 - Setor' : 'Produto'].columns)
emissoes_por_ano = emissoes_gases.melt(id_vars = colunas_info, value_vars = colunas_emissao, var_name = 'Ano', value_name = 'Emissão')
```

```python
emissoes_2021 = emissoes_por_ano[emissoes_por_ano["Ano"] == 2021].groupby('Nível 1 - Setor')[["Emissão"]].mean(numeric_only = True).round(2)
emissoes_2021
```

```python
colunas_ano = list(emissoes_setor.loc[:,1970:2021])
emissoes_setor_total = emissoes_gases.melt(id_vars = "Nível 1 - Setor", value_vars = colunas_ano, var_name = "Ano", value_name = "Emissão")
emissoes_setor_total 
```

```python
emissoes_sum_setor = emissoes_setor_total.groupby("Nível 1 - Setor")["Emissão"].sum().sort_values(ascending = False)
```

```python
emissoes_sum_setor
```

# Exercícios Aula 3

```python
emissoes_setor_ano = emissoes_por_ano.pivot_table(index = "Ano", columns = "Nível 1 - Setor", values = "Emissão", aggfunc= "mean")
emissoes_setor_ano
```

```python
emissoes_setor_ano.plot(subplots = True, figsize = (10,10));
```

```python
# Faça um agrupamento de dados com as colunas "Estado" e "Nível 1 - Setor", obtendo a soma de emissão e armazenando o resultado em uma tabela.

emissao_estado_setor = emissoes_por_ano.groupby(["Estado", "Nível 1 - Setor"])[["Emissão"]].sum()
emissao_estado_setor
```

```python
# Utilizando a tabela construída na atividade 1, selecione os dados referentes à "Energia" do índice "Nível 1 - Setor".

emissao_energia = emissao_estado_setor.xs("Energia", level = 1)
emissao_energia
```

```python
# Utilizando a tabela construída na atividade 1, encontre a atividade econômica com valor máximo de emissão do Estado de Minas Gerais

emissao_estado_setor.xs("MG", level = 0).idxmax()
```

```python
# Obtenha uma tabela contendo a atividade econômica com máxima emissão para cada Estado.

emissao_estado_setor.groupby(level = 0).idxmax()
```

```python
# Obtenha uma tabela contendo o Estado com máxima emissão para cada atividade econômica.

emissao_estado_setor.groupby(level = 1).idxmax()
```
